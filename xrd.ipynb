```python
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from jupyter_jsmol.pymatgen import quick_view
with MPRester("1QkJHLuQRDewfXKS3kq") as m:

    data = m.get_structure_by_material_id("mp-1367")
```

# repr√©sentation graphique des trois premiers pics

```python
XRDCalculator().show_plot(data,two_theta_range=(0,45),annotate_peaks=True,with_labels=True)
```

Le pr√©sent programme calcule les pics d'int√©nsit√© pr√©sents dans le diffractogram du Mg2Si. La source de radiation est le CuK_alpha et la longueur d'onde du rayonnement-x associ√© vaut 1.5444256AÃä (source: https://journals.aps.org/pra/abstract/10.1103/PhysRevA.37.2947). La fonction XRDCalculator() utilise cette longueur d'onde par d√©faut. 

```python
xrdpattern = XRDCalculator().get_pattern(structure = data, scaled = True, two_theta_range = (0,45))
```

# Premier pic
